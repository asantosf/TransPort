```python
import requests
from bs4 import *

travel_date = input('¿Wich year do you want to travel? Type the date in t his format YYYY-MM-DD: ')

spotify_url = f'https://www.billboard.com/charts/hot-100/{travel_date}'

response = requests.get(spotify_url)

soup = BeautifulSoup((response.text))

# Obteniendo lista de nombres de las canciones sin la numero 1, ya que al utilizar la clase, la numero 1 tiene un tipo de letras distinto
song_title = soup.find_all(name='h3', class_='c-title a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 lrv-u-font-size-18@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-330 u-max-width-230@tablet-only')

# Creando lista de canciones sin la numero 1
songs_list = [title.get_text().replace('\n', '').replace('\t', '') for title in song_title]

# obteniendo canción número 1, por tener una clase distinta
top1_song = soup.find(name='h3', class_='c-title a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 u-font-size-23@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-245 u-max-width-230@tablet-only u-letter-spacing-0028@tablet')

# Agregando canción número uno a lista de canciones al principio para no perder el orden del top 100
songs_list.insert(0,top1_song.get_text().replace('\n', '').replace('\t', ''))

number = 1
for song in songs_list:
  print(f'{number}. {song}')
  number+=1
```
